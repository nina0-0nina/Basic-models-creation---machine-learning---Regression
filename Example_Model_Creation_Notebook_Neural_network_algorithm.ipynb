#Imports

```python
import pandas as pd
```

```python
import pickle
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
from sklearn.model_selection import GridSearchCV
```

#Saving the model with pickle

```python
data = [[1, 2, 3], [2, 4, 6], [3, 6, 9], [4, 8, 12], [5, 10, 15], [6, 12, 18], [7, 14, 21], [8, 16, 24], [9, 18, 27], [10, 20, 30]]
df = pd.DataFrame(data, columns=['One', 'Two', 'Three'])
df.head()
```

Data frame separation

---



```python
X = df[['One', 'Two']]
y = df['Three']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_numpy = X_train.to_numpy()
X_test_numpy = X_test.to_numpy()
y_train_numpy = y_train.to_numpy()
y_test_numpy = y_test.to_numpy()
```

#Neural network algorithm

finding the best parameters

---



```python
X_numpy = np.concatenate((X_train_numpy, X_test_numpy), axis = 0)
```

```python
y_numpy = np.concatenate((y_train_numpy, y_test_numpy), axis = 0)
```

```python
params = {'activation':['identity', 'logistic', 'tanh', 'relu'],
              'solver': ['lbfgs', 'sgd', 'adam'],
              'tol': [0.0000100, 0.00000100, 0.0100, 0.00100],
              'hidden_layer_sizes': [(2,2), (3,3), (100)],
          'max_iter': [200000, 150000, 100000, 500000]}
```

```python
grid_search = GridSearchCV(estimator = MLPRegressor(), param_grid = params)
grid_search.fit(X_numpy, y_numpy)
best_params = grid_search.best_params_
best_results = grid_search.best_score_
print(best_params)
print(best_results)
```

Decision tree regressor creation
